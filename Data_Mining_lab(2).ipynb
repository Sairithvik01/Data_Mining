```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
titanic=pd.read_csv('titanic.csv')
print(titanic.head())
print(titanic.info())
print(titanic.describe())
```

```python
sns.set_style("whitegrid")
titanic.hist(figsize=(10, 6), bins=30, edgecolor='red')
plt.suptitle("Histograms of Dataset Features")
plt.show()
```

```python
if not titanic.select_dtypes(include=['object']).empty:
    cat_col = titanic.select_dtypes(include=['object']).columns[0]
    titanic[cat_col].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Bar Chart of {cat_col}")
    plt.xlabel(cat_col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic)
plt.title("Box Plot of Numerical Features")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(titanic, hue='Survived', diag_kind='kde')
plt.title("Pair Plot of Dataset")
plt.suptitle("Pair Plot of Dataset", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
numerical_titanic = titanic.select_dtypes(include=np.number)
sns.heatmap(numerical_titanic.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=titanic)
plt.title("Violin Plot of Numerical Features")
plt.xticks(rotation=45)
plt.show()
```

```python
if not titanic.select_dtypes(include=['object']).empty:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=titanic[cat_col], palette="pastel")
    plt.title(f"Count Plot of {cat_col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
num_cols = titanic.select_dtypes(include=['number']).columns
if len(num_cols) > 1:
    plt.figure(figsize=(10, 6))
    plt.plot(titanic[num_cols[0]], titanic[num_cols[1]], marker="o", linestyle="-", color="b")
    plt.title(f"Line Chart of {num_cols[0]} vs {num_cols[1]}")
    plt.xlabel(num_cols[0])
    plt.ylabel(num_cols[1])
    plt.show()
```
